#### There is interaction between User and Cleverbot.
###### We parse the data into questions and response pairs.

###### We aren't necessarily concerned with who says what, but we are concerned with matching up each response to each question.

We first created a list to hold our
question and response tuples. We then passed our conversations through a function to split
them into these pairs using regular expressions.
Finally, we set it all into a pandas DataFrame with columns labelled q and a.

```python
clist = []
def qa_pairs(x):
    cpairs = re.findall(": (.*?)(?:$|\n)", x)
    clist.extend(list(zip(cpairs, cpairs[1:])))
```

```python
convo.map(qa_pairs);
```

```python
convo_frame = pd.Series(dict(clist)).to_frame().reset_index()
```

```python
convo_frame.columns = ['q', 'a']
```

```python
convo_frame.head()
```

##### Apply a bit of algorithmic magic to match up the closest question to the one a user inputs.


###### Import TfidfVectorization library and the cosine similarity library. 

We then used our training data to create a tf-idf matrix. We can
now use this to transform our own new questions and measure the similarity to existing
questions in our training set.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,3))
```

```python
vec = vectorizer.fit_transform(convo_frame['q'])
```

##### Get the similarity scores

```python
my_q = vectorizer.transform(['Hi. My name is Piush.'])
```

```python
cs = cosine_similarity(my_q, vec)
```

##### This is the cosine similarity between the question I asked and the top five closest questions. To the left is the index and on the right is the cosine similarity.

```python
rs = pd.Series(cs[0]).sort_values(ascending=0)
top5 = rs.iloc[0:5]
top5
```

##### Result output

```python
convo_frame.iloc[top5.index]['q']
```

###### As we can see, nothing is exactly the same, but there are definitely some similarities.

Let's see at the response

```python
rsi = rs.index[120]
rsi
```

```python
convo_frame.iloc[rsi]['a']
```

##### We create a handy function so that we can test a number of statements easily:

```python
def get_response(q):
    my_q = vectorizer.transform([q])
    cs = cosine_similarity(my_q, vec)
    rs = pd.Series(cs[0]).sort_values(ascending=0)
    rsi = rs.index[0]
    return convo_frame.iloc[rsi]['a']
```

```python
get_response("I am feeling sad")
```
