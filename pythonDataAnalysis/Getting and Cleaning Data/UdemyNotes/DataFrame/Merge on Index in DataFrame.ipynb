```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
# for dictionary you set up the column name and the values
dfLeft = DataFrame({'key': ['X','Y','Z','X','Y'],
                    'data':range(5)})
dfLeft
```

```python
dfRight = DataFrame({'groupData': [10,20]},index =['X','Y'])
dfRight
```

```python
#merge with key from the left and index from right 
pd.merge(dfLeft,dfRight,left_on ='key',right_index = True)
```

```python
pd.merge(dfLeft,dfRight,left_on = 'key',right_index = 'groupData')
```

```python
#using a heirarcy index 
dfLeftHr = DataFrame({'key1':['SF','SF','SF','LA','LA'],
                    'key2':[10,20,30,20,30],
                    'dataSet':np.arange(5)})
```

```python
dfLeftHr
```

```python
dfRightHr = DataFrame(np.arange(10).reshape(5,2),
                      index=[['LA','LA','SF','SF','SF'],
                            [20,10,10,10,20]],
                     columns = ['col_1','col_2'])
```

```python
dfRightHr
```

```python
pd.merge(dfLeftHr,dfRightHr,left_on =['key1','key2'],right_index = True)
```

```python
#no need to put so many arguments
#use the join same as merge
dfLeft.join(dfRight)
```
