```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dFrame = DataFrame({'key1': ['A'] *2 + ['B']*3,
                   'key2': [2,2,2,3,3]})
dFrame
```

```python
#how to find duplicates
dFrame.duplicated()
```

```python
#get rid of duplicates
dFrame.drop_duplicates()
```

```python
#filter out the duplicates with columns e.g. dropped by key1
dFrame.drop_duplicates(['key1'])
```

```python
dFrame
```

```python
#by default first value is taken, but you can take last value
dFrame.drop_duplicates(['key1'],take_last =True)
```

```python
dFrame.drop_duplicates(['key1'],keep ='last')
```
