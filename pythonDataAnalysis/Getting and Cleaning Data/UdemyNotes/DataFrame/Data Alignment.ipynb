```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
ser1 = Series([0,1,2],index=['A','B','C'])
ser1
```

```python
ser2 = Series([3,4,5,6],index=['A','B','C','D'])
ser2
```

```python
#add the series together
#adds where the index value matches, replace with a null if doesnot find any
ser1 + ser2
```

```python
#dframe
# to make the columns use the python code list and just add AB
dFrame1 = DataFrame(np.arange(4).reshape(2,2),columns=list('AB'),
                   index =['NYC','LA'])
dFrame1
```

```python
dFrame2 = DataFrame(np.arange(9).reshape((3,3)),columns=list('ADC'),
                   index =['NYC','SF','LA'])
dFrame2
```

```python
#add two dframes
#row index and column match ups are the only one added
dFrame1 + dFrame2
```

```python
#replace the null values
dFrame1
```

```python
dFrame1.add(dFrame2,fill_value = 0)
#only 1 null value, filed in 0 for the 2nd dataframe and kept the original dframe1 values
#important tool for adding different size dataFrame and do not want to lose the values
```
