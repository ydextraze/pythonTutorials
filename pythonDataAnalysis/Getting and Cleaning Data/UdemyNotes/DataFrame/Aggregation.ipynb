```python
# aggregartion result in to scalar such as mean,sum etc.
#Let's get a csv data set to play with
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
```

```python
dFrameWine = pd.read_csv('winequality-red.csv',sep =';')
```

```python
dFrameWine.head()
```

```python
#average alcohol content of the list
dFrameWine['alcohol'].mean()
```

```python
#
dFrameWine
```

```python
#make our own data aggregation method, following is peak to peak function
def max_to_min(arr):
    return arr.max() - arr.min()
```

```python
wino = dFrameWine.groupby('quality')
wino.describe()
```

```python
#apply the own aggreagate function
#general function is done by agg
wino.agg(max_to_min)
```

```python
#string method through aggregrates
wino.agg('mean')
```
